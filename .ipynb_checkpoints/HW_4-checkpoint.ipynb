### Уровень 0:


Посчитайте значение первой и второй производной функции:

$f(x) = x^5 + 4sin(2x) + cos(3x+3)$ 

В точке x = 1.



```python
import numpy as np
from scipy.misc import derivative
def f(x):
    return (x**5)+4*np.sin(2*x)+np.cos(3*x+3)
print('первая производная: ', derivative(f, 1.0, dx=1e-6, n=1))
print('вторая производная: ', derivative(f, 1.0, dx=1e-6, n=2))
```

### Уровень 1:


Постройте график первой и второй производной функции:

$f(x) = sin^5(2x+1)$

На отрезке [-5; 5].

```python
from matplotlib import pyplot as plt

def f2(x):
    return np.sin(2*x+1)**5
def f3(x):
    return derivative(f2, x, dx=1e-6, n=1)
def f4(x):
    return derivative(f2, x, dx=1e-6, n=2)

x_list = np.linspace(-5,5, num=100) 
f3_list = list(map(f3, x_list))
f4_list = list(map(f4, x_list))

plt.plot(x_list,f3_list)
plt.plot(x_list,f4_list)
plt.show()
```
